```python
import pandas as pd
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[-2,3],[-8,7],[2,-2],[8,-9],[7,6],[-3,1]])
y = np.array([0,1,2,3,4,5])#決定是哪一類的
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)#全部Ｘ中的第0個，c=y代表可以隨機取顏色
```

```python
from sklearn.svm import SVC #Classifier
```

```python
clf = SVC()
```

```python
clf.fit(x,y)
```

```python
clf.predict([[-2,3]])
```

```python
clf.predict([[8,9]])
```

```python
clf.predict([[2,-2]])
```

```python
a = np.array( [[i,j] for i in np.arange(-10,4,0.6)  for j in np.arange(-9,4,0.7)])
```

```python
ad = clf.predict(a)
```

```python
plt.scatter(a[:,0],a[:,1],c=ad)#會因為前面的預測影響到後面
```

# K-means

```python
x = np.random.rand(1500,2)
```

```python
plt.scatter(x[:,0],x[:,1],s=50)
```

```python
from sklearn.cluster import KMeans
```

```python
clf = KMeans(n_clusters=10)#開個群組
```

```python
clf.fit(x)#這些參數都是可以調的
```

```python
clf.labels_
```

```python
plt.scatter(x[:,0],x[:,1],c=clf.labels_)
```

```python
clf.predict([[0.5,0.5]])
```

```python
dd = np.array([[i,j]for i in np.arange(-6.3,9.5,1) for j in np.arange(-9.9,0.1,1.8)])
```

```python
ddc = clf.predict(dd)
```

```python
plt.scatter(dd[:,0],dd[:,1],c = ddc)
```
