```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from keras.datasets import mnist
```

```python
(x0_train, y0_train),(x0_test, y0_test) = mnist.load_data()
x_train = x0_train.reshape(x0_train.shape[0], 28, 28, 1)
x_test = x0_test.reshape(x0_test.shape[0], 28, 28, 1)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
```

```python
x_train.shape
```

```python
x_train[87]
```

```python
from keras.utils import np_utils
y_train = np_utils.to_categorical(y0_train,10)#分類帽！！將它分成10種結果
y_test = np_utils.to_categorical(y0_test,10)
```

```python
from keras.models import Sequential
from keras.layers import Dense,Activation, Convolution2D, MaxPooling2D, Dropout, Flatten
from keras.optimizers import SGD
```

```python
model = Sequential()
 
model.add(Convolution2D(32, 3, 3, activation='relu', input_shape=(28, 28, 1)))
model.add(Convolution2D(32, 3, 3, activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
 
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, 
          batch_size=32, epochs=10, verbose=1)
```

```python
score = model.evaluate(x_test, y_test, verbose=0)
```
